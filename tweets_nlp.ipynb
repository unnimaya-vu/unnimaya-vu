NLP

```python
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
tweets=df.text
tweets
```

```python
#tokenization
from nltk import TweetTokenizer#can use word tokenizer or tweet tokenizer
tk=TweetTokenizer()
tweets=tweets.apply(lambda x:tk.tokenize(x)).apply(lambda x:' '.join(x))
tweets
```

Remove special characters

```python
tweets=tweets.str.replace('[^a-zA-Z0-9]+',' ')
tweets
```

To remove the words less than 3

```python
from nltk.tokenize import word_tokenize
tweets=tweets.apply(lambda x:' '.join([w for w in word_tokenize(x) if len(w)>=3]))
tweets
```

Stemming

```python
from nltk.stem import SnowballStemmer
stemmer=SnowballStemmer('english')
tweets=tweets.apply(lambda x:[stemmer.stem(i.lower()) for i in tk.tokenize(x)]).apply(lambda x:' '.join(x))
tweets#output stemmed-all are converted to lower case
```

Remove stopwords

```python
from nltk.corpus import stopwords
sw=stopwords.words('english')
tweets=tweets.apply(lambda x:[i for i in tk.tokenize(x) if i not in sw]).apply(lambda x:' '.join(x))
tweets
```

Vectorization-convert complete data into numerical data

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vec=TfidfVectorizer()
train_data=vec.fit_transform(tweets)
```

```python
train_data.shape
```

```python
print(train_data)
```

```python
y=df['target'].values
y#output to array
```

Model Creation

```python

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(train_data,y,test_size=0.30,random_state=1)
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.metrics import classification_report
k_model=KNeighborsClassifier(n_neighbors=7)
n_model=MultinomialNB()
s_model=SVC()
lst_model=[k_model,n_model,s_model]
```

```python
for i in lst_model:
  print('Model is',i)
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print('*'*50)
  print(confusion_matrix(y_test,y_pred))
  print('Accuracy score is',accuracy_score(y_test,y_pred))
  print('-------Classification Report-----------')
  print(classification_report(y_test,y_pred))
```
