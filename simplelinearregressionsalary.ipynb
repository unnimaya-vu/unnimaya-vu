```python
x=df['YearsExperience']
x
```

```python
y=df['Salary']
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel('year of exp')
plt.ylabel('salary')
plt.title('salary and year of exp graph')
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
x_train
```

```python
#model creation
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred#predicted value
```

```python
y_test#actual value
```

```python
plt.scatter(x_train,y_train,color='r')
#regression line
plt.plot(x_test,y_pred)
```

```python
print(model.predict([[8.5]]))#if year of exp is 8.5 yrs
```

```python
#slope
print('Slope is',model.coef_)
```

```python
#constant/intercept
print('Constant is ',model.intercept_)
```

```python
#error calculation
y_test.tolist()
y_pred.tolist()
df1=pd.DataFrame({'Actual value':y_test,'Predicted value':y_pred,'Difference':y_test-y_pred})
df1
```

```python
#performance evaluation
#mae
from sklearn.metrics import mean_absolute_error
print("Error is",mean_absolute_error(y_test,y_pred))
```

```python
#mape
from sklearn.metrics import mean_absolute_percentage_error
print("Percentage error is",mean_absolute_percentage_error(y_test,y_pred))
```

```python
#mse
from sklearn.metrics import mean_squared_error

mse=mean_squared_error(y_test,y_pred)
print("Squared error is ",mse)
```

```python
#rmse
import numpy as np
sq=np.sqrt(mse)
sq
```

```python
#r2score
from sklearn.metrics import r2_score#coefficient of determination
print("R2_score is ",r2_score(y_test,y_pred))
```
